```python
import matplotlib.pyplot as plt
import random

def islily_num(num):
    digs = [int(d) for d in str(num)]
    for dig in digs:
        if dig == 0 or num % dig != 0:
            return False
    return True

def calculate_lily_prob(length, num_samples):
    lily_count = 0
    for _ in range(num_samples):
        num = random.randint(10**(length-1), 10**length - 1)
        if islily_num(num):
            lily_count += 1
    return lily_count / num_samples

def plot_lily_prob():
    lengths = list(range(2, 20))
    probs = [calculate_lily_prob(length, 1000) for length in lengths]
    
    plt.plot(lengths, probs, marker='o')
    plt.title("Probability of Lily Number vs. Number of Digits")
    plt.xlabel("Number of Digits")
    plt.ylabel("Probability")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_lily_prob()
```

```python
import matplotlib.pyplot as plt

def islily_num(num):
    digs = [int(d) for d in str(num)]
    for dig in digs:
        if dig == 0 or num % dig != 0:
            return False
    return True

def islynch_bell_num(num):
    digs = [int(d) for d in str(num)]
    if len(set(digs)) == len(digs):
        for dig in digs:
            if dig == 0 or num % dig != 0:
                return False
        return True
    return False

def count_lily_nums(dig_length):
    count = 0
    start_range = 10 ** (dig_length - 1)
    end_range = 10 ** dig_length
    for num in range(start_range, end_range):
        if islily_num(num):
            count += 1
    return count

def count_lynch_bell_nums(dig_length):
    count = 0
    start_range = 10 ** (dig_length - 1)
    end_range = 10 ** dig_length
    for num in range(start_range, end_range):
        if islynch_bell_num(num):
            count += 1
    return count

def plot_lily_and_lynch_bell_nums():
    dig_lengths = range(1, 8)
    lily_counts = [count_lily_nums(length) for length in dig_lengths]
    lynch_bell_counts = [count_lynch_bell_nums(length) for length in dig_lengths]

    plt.plot(dig_lengths, lily_counts, label='Lily Numbers', marker='o')
    plt.plot(dig_lengths, lynch_bell_counts, label='Lynch Bell Numbers', marker='s')
    plt.xlabel('Digit Length')
    plt.ylabel('Amount of Numbers')
    plt.title('Lily Numbers vs Lynch Bell Numbers')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_lily_and_lynch_bell_nums()
```
